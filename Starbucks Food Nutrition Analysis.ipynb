```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import style
style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'starbucks-menu-nutrition-food.csv', encoding = 'utf - 16')
data.head(20)
```

```python
data.info()
```

```python
data.rename(columns = {'Unnamed: 0':'Food items'}, inplace =True)
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.columns = data.columns.str.lstrip()
data.columns = data.columns.str.rstrip()
```

```python
print(data['Protein (g)'].mean())
print(data['Calories'].mean())
print(data['Fat (g)'].mean())
print(data['Carb. (g)'].mean())
print(data['Fiber (g)'].mean())
```

```python
high_Protein = data[data['Protein (g)']>15]
low_Protein = data[data['Protein (g)']==min(data['Protein (g)'])]
high_Calories = data[data['Calories']>360]
low_Calories = data[data['Calories']==min(data['Calories'])]
high_Fat = data[data['Fat (g)']>17]
low_Fat = data[data['Fat (g)']== min(data['Fat (g)'])]
high_carbs = data[data['Carb. (g)']>42]
low_carbs = data[data['Carb. (g)']==min(data['Carb. (g)'])]
high_fiber = data[data['Fiber (g)']>3]
low_fiber = data[data['Fiber (g)']==min(data['Fiber (g)'])]
```

```python
plt.figure(figsize = (8,8), dpi = 200)
sns.barplot('Protein (g)','Food items', data = high_Protein )
```

```python
plt.figure(figsize = (14,2), dpi = 200)
sns.barplot('Food items','Protein (g)', data = low_Protein )
plt.xticks(rotation = 15)
```

#### Turkey Pesto Panini and 'Roasted Turkey & Dill Havarti Sandwich' is the only food with high protein content whereas Frappuccino Cookie Straw, Birthday Cake Pop, Pirate Cake Pop, Organic Avocado (Spread), Organic Avocado (Spread) having the low protein content. 

```python
plt.figure(figsize = (8,10), dpi = 200)
sns.barplot('Calories','Food items', data = high_Calories )
```

```python
plt.figure(figsize = (12,3), dpi = 200, facecolor = '#d8e2f2')
sns.barplot('Food items','Calories', data = low_Calories)
```

#### Lentils & Vegetable Protein Bowl with Brown Rice is having higest calorie where as Frappuccino Cookie Straw, Organic Avocado (Spread),Seasonal Fruit Blend is low calorie food

```python
plt.figure(figsize = (8,10), dpi = 200, facecolor = '#d8e2f2')
sns.barplot('Fat (g)','Food items', data = high_Fat)
```

```python
plt.figure(figsize = (10,6), dpi = 200, facecolor = '#d8e2f2')
sns.barplot('Fiber (g)','Food items', data = high_fiber)
```

#### Green Goddess Avocado Salad is having the higest fat level and Lentils & Vegetable Protein Bowl with Brown Rice is having higher fiber content

```python
plt.figure(figsize = (12,10), dpi = 200, facecolor = '#d8e2f2')
sns.barplot('Carb. (g)','Food items', data = high_carbs)
```

```python
plt.figure(figsize = (16,4), dpi = 200, facecolor = '#d8e2f2')
sns.barplot('Food items','Carb. (g)', data = low_carbs)
```

#### 'Strawberries & Jam Sandwich' and 'Lentils & Vegetable Protein Bowl with Brown Rice' is having the higest carbs whereas Organic Avocado (Spread) Organic is having the least carb content Avocado (Spread) is having the least carb content

```python
plt.figure(figsize = (16,4), dpi = 200, facecolor = '#d8e2f2')
plt.subplot(2,2,1)
sns.scatterplot('Calories','Fat (g)', data = data)
plt.subplot(2,2,2)
sns.scatterplot('Calories','Carb. (g)', data = data)
plt.subplot(2,2,3)
sns.scatterplot('Calories','Fiber (g)', data = data)
plt.subplot(2,2,4)
sns.scatterplot('Calories','Protein (g)', data = data)
```

#### With increase in fat, carb, protein intake calories increases whereas no major effect seen with increase in fiber intake

```python
plt.figure(figsize = (14,4), dpi = 200, facecolor = '#d8e2f2')
plt.subplot(3,2,1)
sns.scatterplot('Fat (g)','Carb. (g)', data = data)
plt.tight_layout()
plt.subplot(3,2,2)
sns.scatterplot('Fat (g)','Fiber (g)', data = data)
plt.subplot(3,2,3)
sns.scatterplot('Fat (g)', 'Protein (g)',data = data)
plt.subplot(3,2,4)
sns.scatterplot('Protein (g)','Carb. (g)', data = data)
plt.subplot(3,2,5)
sns.scatterplot('Carb. (g)', 'Fiber (g)',data = data)
plt.subplot(3,2,6)
sns.scatterplot('Protein (g)','Fiber (g)',data = data)
```

#### Food with increase in fat, carb and protein content no changes found in fiber content.
#### Food with increase in carb no changes found in fat and protein content.
#### Food with high protein content may have increase fat content.
