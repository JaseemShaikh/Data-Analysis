```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import style
style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('starbucks-menu-nutrition-drinks.csv')
data.head(20)
```

```python
data.info()
```

```python
data.replace('-', np.nan, inplace = True)
```

```python
data.head(20)
```

```python
data['Calories'] = pd.to_numeric(data['Calories']) 
data['Fat (g)'] = pd.to_numeric(data['Fat (g)']) 
data['Carb. (g)'] = pd.to_numeric(data['Carb. (g)']) 
data['Fiber (g)'] = pd.to_numeric(data['Fiber (g)']) 
data['Protein'] = pd.to_numeric(data['Protein']) 
data['Sodium'] = pd.to_numeric(data['Sodium']) 
```

```python
print(data['Calories'].dtypes)
print(data['Fat (g)'].dtypes)
print(data['Carb. (g)'].dtypes)
print(data['Fiber (g)'].dtypes)
print(data['Protein'].dtypes)
print(data['Sodium'].dtypes)
```

```python
data[data.isna().any(axis=1)]
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
data.rename(columns = {'Unnamed: 0':'Beverage_category'}, inplace = True)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.duplicated().sum()
```

```python
data.head(20)
```

```python
high_calorie = data[data['Calories']>150]
high_fat = data[data['Fat (g)']>5]
high_carb = data[data['Carb. (g)']>30]
high_fiber = data[data['Fiber (g)']>1]
high_protein = data[data['Protein']>5]
high_sodium = data[data['Sodium']>100]
```

```python
plt.figure(figsize = (12, 6), dpi = 200)
sns.barplot('Calories','Beverage_category', data = high_calorie)
```

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.barplot('Fat (g)','Beverage_category', data = high_fat)
```

#### Among all the high calorie and high fat content beverage (above average) 'Starbucks signature hot chocolate' has the highest calorie and fat content.

```python
plt.figure(figsize = (12, 7), dpi = 200)
sns.barplot('Carb. (g)','Beverage_category', data = high_carb)
```

#### Among all the high carb beverage  'Cinnamon Dolce Frappuccino Blended Coffee' is having high carb content.

```python
plt.figure(figsize = (12, 4), dpi = 200)
sns.barplot('Fiber (g)','Beverage_category', data = high_fiber)
```

#### Among all the high fiber beverage 'Chocolate Smoothie' is having high fiber content. 

```python
plt.figure(figsize = (12, 6), dpi = 200)
sns.barplot('Protein','Beverage_category', data = high_protein)
```

#### Among all the high protein beverage 'Chocolate Smoothie' ,'Starbucks Doubleshot Protein Vanilla','Starbucks Doubleshot Protein Dark chocolate' is having high protein content. 

```python
plt.figure(figsize = (12, 6), dpi = 200)
sns.barplot('Sodium','Beverage_category', data = high_sodium)
```

#### 'White Chocolate Mocha' is having the higest sodium content.

```python
plt.figure(figsize = (16,4), dpi = 200, facecolor = '#d8e2f2')
plt.subplot(2,2,1)
sns.scatterplot('Calories','Fat (g)', data = data)
plt.subplot(2,2,2)
plt.tight_layout()
sns.scatterplot('Calories','Carb. (g)', data = data)
plt.subplot(2,2,3)
sns.scatterplot('Calories','Fiber (g)', data = data)
plt.subplot(2,2,4)
sns.scatterplot('Calories','Protein', data = data)

plt.figure(figsize = (20,4), dpi = 200, facecolor = '#d8e2f2')
plt.subplot(1,1,1)
sns.scatterplot('Calories','Sodium', data = data)
```

#### With increase in fat, carb, protein and sodium in beverages calories increases whereas no major effect seen in calories with increase in fiber intake

```python
plt.figure(figsize = (16,8), dpi = 200)
plt.subplot(5,2,1)
sns.scatterplot('Carb. (g)','Fat (g)', data = data)
plt.subplot(5,2,2)
sns.scatterplot('Fat (g)','Fiber (g)', data = data)
plt.subplot(5,2,3)
sns.scatterplot('Sodium','Protein', data = data)
plt.subplot(5,2,4)
sns.scatterplot('Fat (g)','Protein', data = data)
plt.subplot(5,2,5)
sns.scatterplot('Fat (g)','Sodium', data = data)
plt.subplot(5,2,6)
sns.scatterplot('Carb. (g)','Fiber (g)', data = data)
plt.subplot(5,2,7)
sns.scatterplot('Carb. (g)','Protein', data = data)
plt.subplot(5,2,8)
sns.scatterplot('Carb. (g)','Sodium', data = data)
plt.subplot(5,2,9)
sns.scatterplot('Protein','Fiber (g)', data = data)
plt.subplot(5,2,10)
sns.scatterplot('Sodium','Fiber (g)', data = data)
plt.tight_layout()
```

#### Beverage with high protein content also have increased fat content.
